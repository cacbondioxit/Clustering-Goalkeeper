```python
df1 = data1.merge(data2, on='Player', how='inner')

df1
```

```python
df1.columns
```

```python
# 필요없는 행 삭제 (골키퍼의 실점은 삭제하려다가 결국 데이터에 포함)

df = df1.drop(['Rk_x', 'Player', 'Nation_x', 'Pos_x', 'Squad_x', 'Age_x', 'Born_x', 'MP', 'Starts', 'Min', '90s_x',  
               # 'GA_x', 'GA90', 
               'W', 'D', 'L', 'Rk_y', 'Nation_y', 'Pos_y', 'Squad_y', 'Age_y', 'Born_y', '90s_y', 'GA_y', 'PKA_y'], axis=1)

df
```

```python
df.dtypes
```

```python
# 가능한 다른 년도 데이터도 추가할 예정
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df)

df_pca
```

```python
df_pca = pd.DataFrame(df_pca, columns=['pca1', 'pca2'])

df_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
# 주성분의 3개만 써도 전체의 97%나 설명된다.
```

```python
sns.scatterplot(x="pca1", y="pca2", data=df_pca)
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df)
    distortions.append(kmeanModel.inertia_)
plt.figure(figsize=(12,4))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeanModel = KMeans(n_clusters=2)
kmeanModel.fit(df)
df_pca['k_means'] = kmeanModel.predict(df)

```

```python
fig, ax = plt.subplots(figsize=(18,12))
ax.scatter(df_pca["pca1"], df_pca["pca2"], c=df_pca['k_means'], cmap=plt.cm.Set1)
```

```python
df_res = pd.concat([df1['Player'], df_pca['k_means']], axis=1)
```
